```python
import numpy as np
import matplotlib.pyplot as plt

iris = np.genfromtxt("iris.txt",delimiter=None) # load the data
Y = iris[:,-1]
X = iris[:,0:2]

import mltools as ml

X,Y = ml.shuffleData(X,Y);
Xtr,Xva,Ytr,Yva = ml.splitData(X, Y, 0.75);


```

```python
print("k = 1")
knn = ml.knn.knnClassify()
knn.train(Xtr, Ytr, K=1)
ml.plotClassify2D(knn, Xtr, Ytr)
plt.show()
```

```python
print("k = 5")
knn = ml.knn.knnClassify()
knn.train(Xtr, Ytr, K=5)
ml.plotClassify2D(knn, Xtr, Ytr)
plt.show()
```

```python
print("k = 10")
knn = ml.knn.knnClassify()
knn.train(Xtr, Ytr, K=10)
ml.plotClassify2D(knn, Xtr, Ytr)
plt.show()
```

```python
print("k = 50")
knn = ml.knn.knnClassify()
knn.train(Xtr, Ytr, K=50)
ml.plotClassify2D(knn, Xtr, Ytr)
plt.show()
```

```python
print("The error rate for the first two features")
K=[1,2,5,10,50,100,200];
train_err = [0,0,0,0,0,0,0]
val_err =   [0,0,0,0,0,0,0]
X,Y = ml.shuffleData(X,Y);
Xtr,Xva,Ytr,Yva = ml.splitData(X, Y, 0.75);
for i,k in enumerate(K):
    learner = ml.knn.knnClassify()
    learner.train(Xtr, Ytr, k)
    YhatTrain = learner.predict(Xtr)
    YhatValid = learner.predict(Xva)
    train_err[i] = learner.err(Xtr, Ytr)
    val_err[i]   = learner.err(Xva, Yva)
    
plt.semilogx( train_err, 'r')
plt.semilogx( val_err, 'g')
plt.show()
```

```python
print("The error rate for all of the features")
Y = iris[:,-1]
X = iris[:,0:-1]
X,Y = ml.shuffleData(X,Y);
Xtr,Xva,Ytr,Yva = ml.splitData(X, Y, 0.75);
K=[1,2,5,10,50,100,200];
train_err = [0,0,0,0,0,0,0]
val_err =   [0,0,0,0,0,0,0]
X,Y = ml.shuffleData(X,Y);
Xtr,Xva,Ytr,Yva = ml.splitData(X, Y, 0.75);
for i,k in enumerate(K):
    learner = ml.knn.knnClassify()
    learner.train(Xtr, Ytr, k)
    YhatTrain = learner.predict(Xtr)
    YhatValid = learner.predict(Xva)
    train_err[i] = learner.err(Xtr, Ytr)
    val_err[i]   = learner.err(Xva, Yva)
plt.semilogx( train_err, 'r')
plt.semilogx( val_err, 'g')
plt.show()
```

```python
p(y = 1) = 4 / 10 
p(y = -1) = 6 / 10
P(x1 = 1 | y = 1) = 3 / 4  P(x1 = 0 | y = 1) = 1 / 4 
P(x2 = 1 | y = 1) = 0 / 4  P(x1 = 0 | y = 1) = 4 / 4 
P(x3 = 1 | y = 1) = 3 / 4  P(x1 = 0 | y = 1) = 1 / 4 
P(x4 = 1 | y = 1) = 2 / 4  P(x1 = 0 | y = 1) = 2 / 4 
p(x5 = 1 | y = 1) = 1 / 4  P(x1 = 0 | y = 1) = 3 / 4


P(x1 = 1 | y = -1) = 3 / 6  P(x1 = 0 | y = -1) = 3 / 6 
P(x2 = 1 | y = -1) = 5 / 6  P(x1 = 0 | y = -1) = 1 / 6 
P(x3 = 1 | y = -1) = 4 / 6  P(x1 = 0 | y = -1) = 2 / 6 
P(x4 = 1 | y = -1) = 5 / 6  P(x1 = 0 | y = -1) = 1 / 6 
p(x5 = 1 | y = -1) = 2 / 6  P(x1 = 0 | y = -1) = 4 / 6
```
