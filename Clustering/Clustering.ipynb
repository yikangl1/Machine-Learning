```python
# 1.1

import numpy as np
import matplotlib.pyplot as plt
import mltools as ml
import scipy.linalg

iris = np.genfromtxt('iris.txt', delimiter=None)

X, Y = iris[:,0:2], iris[:,-1]

plt.scatter(X[:,0],X[:,1],color='b')
plt.xlabel('feature x1')
plt.ylabel('feature x2')
plt.title('Clustered algorithm graph')
plt.show()
```

```python
# 1.2

z,c,d = ml.cluster.kmeans(X,2)
ml.plotClassify2D(None, X, z)
plt.title('K = 2')
plt.xlabel('feature x1')
plt.ylabel('feature x2')
plt.show()

z,c,d = ml.cluster.kmeans(X,5)
ml.plotClassify2D(None, X, z)
plt.title('K = 5')
plt.xlabel('feature x1')
plt.ylabel('feature x2')
plt.show()

z,c,d = ml.cluster.kmeans(X,20)
ml.plotClassify2D(None, X, z)
plt.title('K = 20')
plt.xlabel('feature x1')
plt.ylabel('feature x2')
plt.show()
```

```python
# 1. 3

z, c = ml.cluster.agglomerative(X, 2, method='min')
plt.title("Single Linkage for K = 2");
ml.plotClassify2D(None, X, z);
plt.show()

z, c = ml.cluster.agglomerative(X, 2, method='max')
plt.title("Complete Linkage for K = 2");
ml.plotClassify2D(None, X, z);
plt.show()

z, c = ml.cluster.agglomerative(X, 5, method='min')
plt.title("Single Linkage for K = 5");
ml.plotClassify2D(None, X, z);
plt.show()

z, c = ml.cluster.agglomerative(X, 5, method='max')
plt.title("Complete Linkage for K = 5");
ml.plotClassify2D(None, X, z);
plt.show()

z, c = ml.cluster.agglomerative(X, 20, method='min')
plt.title("Single Linkage for K = 20");
ml.plotClassify2D(None, X, z);
plt.show()

z, c = ml.cluster.agglomerative(X, 20, method='max')
plt.title("Complete Linkage for K = 20");
ml.plotClassify2D(None, X, z);
plt.show()
```

```python
# 1.4 

'''
 The difference between agglomerative clusters and k-means  is that
agglomerative clusters are dendograms. When we use minimum distance 
between clusters it will produce a minimum spanning tree while a
maximum distance will avoid elongated clusters. This is shown
in the single and complete linkage for each as single linkage has
a few clusters that take up the majority while the rest are small or
single nodes. K-means base each cluster on a center point. The
initialization of each center may change how the clusters look. 
Distance based or random.
'''
```

```python
# 2.1

X = np.genfromtxt("faces.txt", delimiter=None) # load face dataset
M, N = X.shape
plt.figure()
# pick a data point i for display
img = np.reshape(X[5,:],(24,24)) # convert vectorized data point to 24x24 image patch
plt.imshow( img.T , cmap="gray") # display image patch; you may have to squint
plt.show()

# 2A
mean = np.mean(X)
X0 = X-mean

print("X0 = ",X0)
```

```python
# 2.2 

U, S, V = scipy.linalg.svd(X0, full_matrices=False)
W = U.dot(np.diag(S))

print(W.shape)
print(V.shape)
```

```python
# 2.3

mse = []
for k in range(1, 11):
    X0hat = W[:, :k].dot(V[:k,:])
    mse.append(np.mean((X0 - X0hat)**2))
# plot the data
_, axis = plt.subplots()
axis.plot(range(1,11), mse, c='red')
axis.set_xticks(range(1,11))
plt.show()
```

```python
# 2.4

mu = np.mean(X, axis=0)
X0 = np.array([x - mu for x in X])

U, s, Vh = scipy.linalg.svd(X0)
S = scipy.linalg.diagsvd(s, U.shape[1], Vh.shape[0])
W = U.dot(S)

for j in range(0, 3):
    a = 2 * np.median(np.abs(W[:,j])) #scale factor
    
    p0 = mu - a * Vh[j,:]
    p1 = mu + a * Vh[j,:]
    
    img0 = np.reshape(p0, (24, 24))
    img1 = np.reshape(p1, (24, 24))
    
    print("Negative direction  " + str(j + 1))
    plt.imshow(img0.T, cmap='gray')
    plt.show()
    
    print("Positive direction  " + str(j + 1))
    plt.imshow(img1.T, cmap='gray')
    plt.show()
```

```python
# 2.5

faceHats = []
K = [5, 10, 50, 100]
for k in K:
    faceHat = np.dot(W[:,:k], Vh[:k,:])
    faceHats.append(faceHat)

faces = np.random.randint(0, M, 2)

for face in faces:
    for i in range(0, len(K)):
        print("Approximation of face #" + str(face) + " where k=" + str(K[i]))
        
        img = np.reshape(faceHats[i][face,:], (24, 24))
        plt.imshow(img.T, cmap="gray")
        plt.show()
    
    print("Original image for face # " + str(face))
    img = np.reshape(X[face,:], (24, 24))
    plt.imshow(img.T, cmap="gray")
    plt.show()
```
